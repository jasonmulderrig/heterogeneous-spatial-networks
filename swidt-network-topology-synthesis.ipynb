```python
# This may or may not correspond to the number of cpus for optimal
# parallelization performance. Feel free to modify if you see fit.
cpu_num = int(np.floor(multiprocessing.cpu_count()/2))

# Initialization of identification information for this particular batch
# of spider web-inspired Delaunay-triangulated networks
date = "20240815"
batch = "A"
network = "swidt"
scheme = "rccs"

date_str = f"date_{date}"
batch_str = f"batch_{batch}"
network_str = f"network_{network}"
scheme_str = f"scheme_{scheme}"

dim_str = "dim"
b_str = "b"
n_str = "n"
k_str = "k"
eta_n_str = "eta_n"
config_str = "config"
pruning_str = "pruning"

filepath = filepath_str(network)
filename_prefix = filepath + f"{date}{batch}"

identifier_filename = filename_prefix + "-identifier" + ".txt"
dim_filename = filename_prefix + f"-{dim_str}" + ".dat"
b_filename = filename_prefix + f"-{b_str}" + ".dat"
n_filename = filename_prefix + f"-{n_str}" + ".dat"
k_filename = filename_prefix + f"-{k_str}" + ".dat"
eta_n_filename = filename_prefix + f"-{eta_n_str}" + ".dat"
config_filename = filename_prefix + f"-{config_str}" + ".dat"
pruning_filename = filename_prefix + f"-{pruning_str}" + ".dat"
params_filename = filename_prefix + "-params" + ".dat"
sample_params_filename = filename_prefix + "-sample_params" + ".dat"
sample_config_params_filename = (
    filename_prefix + "-sample_config_params" + ".dat"
)
sample_config_pruning_params_filename = (
    filename_prefix + "-sample_config_pruning_params" + ".dat"
)

identifier_arr = np.asarray(
    [date_str, batch_str, network_str, scheme_str, dim_str, b_str, n_str,
     k_str, eta_n_str])

# Initialization of fundamental parameters for spider web-inspired
# Delaunay-triangulated networks
dim_arr = np.asarray([2, 3], dtype=int)
b_arr = np.asarray([1.0])
n_arr = np.asarray([100, 150, 224, 334, 500, 748, 1118, 1672, 2500], dtype=int) # = np.rint(np.logspace(np.log10(100), np.log10(2500), num=9)).astype(int)
k_arr = np.arange(3, 9, dtype=int) # 3, 4, 5, 6, 7, 8
eta_n_arr = np.asarray([0.025, 0.05, 0.075, 0.1])
config_arr = np.arange(4, dtype=int) # 0, 1, 2, 3
pruning_arr = np.arange(3, dtype=int) # 0, 1, 2

dim_num = np.shape(dim_arr)[0]
b_num = np.shape(b_arr)[0]
n_num = np.shape(n_arr)[0]
k_num = np.shape(k_arr)[0]
eta_n_num = np.shape(eta_n_arr)[0]
config_num = np.shape(config_arr)[0]
pruning_num = np.shape(pruning_arr)[0]

sample_num = dim_num * b_num * n_num * k_num * eta_n_num
sample_config_num = sample_num * config_num
sample_config_pruning_num = sample_num * config_num * pruning_num

# Populate the network parameters array
params_arr = np.empty([sample_num, 5])
sample = 0
for dim in np.nditer(dim_arr):
    for b in np.nditer(b_arr):
        for n in np.nditer(n_arr):
            for k in np.nditer(k_arr):
                for eta_n in np.nditer(eta_n_arr):
                    params_arr[sample, :] = np.asarray([dim, b, n, k, eta_n])
                    sample += 1

# Populate the network sample parameters array
sample_params_arr = np.empty([sample_num, 6])
sample = 0
for dim in np.nditer(dim_arr):
    for b in np.nditer(b_arr):
        for n in np.nditer(n_arr):
            for k in np.nditer(k_arr):
                for eta_n in np.nditer(eta_n_arr):
                    sample_params_arr[sample, :] = (
                        np.asarray([sample, dim, b, n, k, eta_n])
                    )
                    sample += 1

# Populate the network sample configuration parameters array
sample_config_params_arr = np.empty([sample_config_num, 7])
sample = 0
indx = 0
for dim in np.nditer(dim_arr):
    for b in np.nditer(b_arr):
        for n in np.nditer(n_arr):
            for k in np.nditer(k_arr):
                for eta_n in np.nditer(eta_n_arr):
                    for config in np.nditer(config_arr):
                        sample_config_params_arr[indx, :] = (
                            np.asarray([sample, dim, b, n, k, eta_n, config])
                        )
                        indx += 1
                    sample += 1

# Populate the network sample configuration pruning parameters array
sample_config_pruning_params_arr = np.empty([sample_config_pruning_num, 8])
sample = 0
indx = 0
for dim in np.nditer(dim_arr):
    for b in np.nditer(b_arr):
        for n in np.nditer(n_arr):
            for k in np.nditer(k_arr):
                for eta_n in np.nditer(eta_n_arr):
                    for config in np.nditer(config_arr):
                        for pruning in np.nditer(pruning_arr):
                            sample_config_pruning_params_arr[indx, :] = (
                                np.asarray([sample, dim, b, n, k, eta_n, config, pruning])
                            )
                            indx += 1
                    sample += 1

# Save identification information and fundamental network parameters
np.savetxt(identifier_filename, identifier_arr, fmt="%s")
np.savetxt(dim_filename, dim_arr, fmt="%d")
np.savetxt(b_filename, b_arr)
np.savetxt(n_filename, n_arr, fmt="%d")
np.savetxt(k_filename, k_arr, fmt="%d")
np.savetxt(eta_n_filename, eta_n_arr)
np.savetxt(config_filename, config_arr, fmt="%d")
np.savetxt(pruning_filename, pruning_arr, fmt="%d")
np.savetxt(params_filename, params_arr)
np.savetxt(sample_params_filename, sample_params_arr)
np.savetxt(sample_config_params_filename, sample_config_params_arr)
np.savetxt(sample_config_pruning_params_filename, sample_config_pruning_params_arr)
```

```python
# Calculate and save L for each swidt parameter sample
from heterogeneous_spatial_networks_worker import run_swidt_L

swidt_L_params_arr = (
    sample_params_arr[:, [0, 1, 2, 3, 5]]
) # sample, dim, b, n, eta_n
swidt_L_params_list = list(map(tuple, swidt_L_params_arr))
swidt_L_args = (
    [
        (network, date, batch, int(sample), int(dim), b, int(n), eta_n)
        for (sample, dim, b, n, eta_n) in swidt_L_params_list
    ]
)

__spec__ = None
with multiprocessing.Pool(processes=cpu_num) as pool:
    pool.map(run_swidt_L, swidt_L_args)
```

```python
# Perform the random cross-linker seeding procedure for each swidt
# parameter sample configuration
from heterogeneous_spatial_networks_worker import run_crosslinker_seeding

max_try = 100

crosslinker_seeding_params_arr = (
    sample_config_params_arr[:, [0, 1, 2, 3, 6]]
) # sample, dim, b, n, config
crosslinker_seeding_params_list = (
    list(map(tuple, crosslinker_seeding_params_arr))
)
crosslinker_seeding_args = (
    [
        (network, date, batch, int(sample), scheme, int(dim), b, int(n), int(config), int(max_try))
        for (sample, dim, b, n, config) in crosslinker_seeding_params_list
    ]
)

__spec__ = None
with multiprocessing.Pool(processes=cpu_num) as pool:
    pool.map(run_crosslinker_seeding, crosslinker_seeding_args)
```

```python
# Check to see if the number of cross-linkers seeded in the random
# cross-linker procedure, rccs_n, equals the intended/specified number
# of cross-linkers to be seeded, n. Continue to the topology
# initialization procedure ONLY IF rccs_n = n. If rccs_n != n for any
# specified network, then the code block identifies which particular
# set(s) of network parameters rccs_n != n occurred for.
if scheme == "rccs":
    rccs_n_vs_n = np.zeros(sample_config_num)
    for indx in range(sample_config_num):
        sample = int(sample_config_params_arr[indx, 0])
        n = int(sample_config_params_arr[indx, 3])
        config = int(sample_config_params_arr[indx, 6])
        
        rccs_filename = (
            filename_str(network, date, batch, sample)
            + f"C{config:d}" + ".config"
        )
        rccs = np.loadtxt(rccs_filename)
        
        if np.shape(rccs)[0] == n:
            rccs_n_vs_n[indx] = 1
        else: pass

    sample_config_params_rccs_n_neq_n = (
        sample_config_params_arr[np.where(rccs_n_vs_n == 0)]
    )
    
    if np.shape(sample_config_params_rccs_n_neq_n)[0] == 0:
        print_str = (
            "Success! rccs_n = n from the random core cross-linker seeding "
            + "procedure for all network parameters!"
        )
        print(print_str)
    elif np.shape(sample_config_params_rccs_n_neq_n)[0] > 0:
        print_str = (
            "rccs_n != n from the random core cross-linker seeding "
            + "procedure for at least one set of network parameters. "
            + "Repeat the random core cross-linker seeding procedure "
            + "for the applicable set of network parameters before "
            + "continuing on to the topology initialization procedure."
        )
        print(print_str)
```
