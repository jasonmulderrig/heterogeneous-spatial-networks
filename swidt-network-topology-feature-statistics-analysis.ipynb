```python
# This may or may not correspond to the number of cpus for optimal
# parallelization performance. Feel free to modify if you see fit.
cpu_num = int(np.floor(multiprocessing.cpu_count()/2))

# Initialization of identification information for this particular batch
# of spider web-inspired Delaunay-triangulated networks
date = "20240815"
batch = "A"
network = "swidt"
scheme = "rccs"

core_pb = "core_pb"
core_pb_conn = "core_pb_conn"

filepath = filepath_str(network)
filename_prefix = filepath + f"{date}{batch}"

sample_config_pruning_params_filename = (
    filename_prefix + "-sample_config_pruning_params" + ".dat"
)
sample_config_pruning_params_arr = np.loadtxt(sample_config_pruning_params_filename)
sample_config_pruning_num = np.shape(sample_config_pruning_params_arr)[0]

sample_config_pruning_arr = (
    sample_config_pruning_params_arr[:, [0, 6, 7]]
) # sample, config, pruning
sample_config_pruning_list = list(map(tuple, sample_config_pruning_arr))
sample_config_pruning_core_pb_args = (
    [
        (network, date, batch, int(sample), int(config), int(pruning), core_pb)
        for (sample, config, pruning) in sample_config_pruning_list
    ]
)
sample_config_pruning_core_pb_conn_args = (
    [
        (network, date, batch, int(sample), int(config), int(pruning), core_pb_conn)
        for (sample, config, pruning) in sample_config_pruning_list
    ]
)

sample_l_bound_config_pruning_arr = (
    sample_config_pruning_params_arr[:, [0, 4, 6, 7]]
) # sample, k, config, pruning
# k -> l_bound
for sample in range(np.shape(sample_l_bound_config_pruning_arr)[0]):
    if sample_l_bound_config_pruning_arr[sample, 1] == 3:
        sample_l_bound_config_pruning_arr[sample, 1] = 25
    elif sample_l_bound_config_pruning_arr[sample, 1] == 4:
        sample_l_bound_config_pruning_arr[sample, 1] = 15
    elif sample_l_bound_config_pruning_arr[sample, 1] == 5:
        sample_l_bound_config_pruning_arr[sample, 1] = 14
    elif sample_l_bound_config_pruning_arr[sample, 1] == 6:
        sample_l_bound_config_pruning_arr[sample, 1] = 12
    elif sample_l_bound_config_pruning_arr[sample, 1] == 7:
        sample_l_bound_config_pruning_arr[sample, 1] = 10
    elif sample_l_bound_config_pruning_arr[sample, 1] == 8:
        sample_l_bound_config_pruning_arr[sample, 1] = 10
sample_l_bound_config_pruning_list = list(map(tuple, sample_l_bound_config_pruning_arr))
sample_l_bound_config_pruning_core_pb_args = (
    [
        (network, date, batch, int(sample), int(l_bound), int(config), int(pruning), core_pb)
        for (sample, l_bound, config, pruning) in sample_l_bound_config_pruning_list
    ]
)
sample_l_bound_config_pruning_core_pb_conn_args = (
    [
        (network, date, batch, int(sample), int(l_bound), int(config), int(pruning), core_pb_conn)
        for (sample, l_bound, config, pruning) in sample_l_bound_config_pruning_list
    ]
)

sample_dim_config_pruning_arr = (
    sample_config_pruning_params_arr[:, [0, 1, 6, 7]]
) # sample, dim, config, pruning
sample_dim_config_pruning_list = list(map(tuple, sample_dim_config_pruning_arr))
sample_dim_config_pruning_core_pb_args = (
    [
        (network, date, batch, int(sample), int(dim), int(config), int(pruning), core_pb)
        for (sample, dim, config, pruning) in sample_dim_config_pruning_list
    ]
)
sample_dim_config_pruning_core_pb_conn_args = (
    [
        (network, date, batch, int(sample), int(dim), int(config), int(pruning), core_pb_conn)
        for (sample, dim, config, pruning) in sample_dim_config_pruning_list
    ]
)
```

```python
# Calculate the degree of each cross-linker node in each swidt sample, and bin this information.
# Perform this calculation for the core_pb graph and the core_pb_conn 
# graph.
from heterogeneous_spatial_networks_worker import run_swidt_network_graph_k_counts

__spec__ = None
with multiprocessing.Pool(processes=cpu_num) as pool:
    pool.map(run_swidt_network_graph_k_counts, sample_config_pruning_core_pb_args)
    pool.map(run_swidt_network_graph_k_counts, sample_config_pruning_core_pb_conn_args)
```

```python
# Calculate the chordless cycles in each swidt sample, and bin this information.
# Perform this calculation for the core_pb graph and the core_pb_conn 
# graph.
from heterogeneous_spatial_networks_worker import run_swidt_network_graph_h_counts

__spec__ = None
with multiprocessing.Pool(processes=cpu_num) as pool:
    pool.map(run_swidt_network_graph_h_counts, sample_l_bound_config_pruning_core_pb_args)
    pool.map(run_swidt_network_graph_h_counts, sample_l_bound_config_pruning_core_pb_conn_args)
```
